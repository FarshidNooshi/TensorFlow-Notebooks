# Plotting the data

```python
fig, ax = plt.subplots(figsize=(24, 16), nrows=3, ncols=7)

cnt = 0
for i in range(7):
    for j in range(i + 1, 7):
        ax[cnt // 7, cnt % 7].scatter(df[df.columns[i]], df[df.columns[j]], c=df['class'], cmap='viridis', s=20)
        ax[cnt // 7, cnt % 7].set_xlabel(df.columns[i])
        ax[cnt // 7, cnt % 7].set_ylabel(df.columns[j])
        ax[cnt // 7, cnt % 7].set_title('{} vs {}'.format(df.columns[i], df.columns[j]))
        cnt += 1

fig.tight_layout()
        
```
